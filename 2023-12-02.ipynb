```python
# Part One

limits = {
    'red': 12,
    'green': 13,
    'blue': 14,
}

game_sum = 0

with open("2023-12-02_puzzle_input.txt") as f:
    for game in f:
        if game is None:
            continue
        game = game.strip()
        raw_game, raw_rounds = game.split(": ")
        _, game_number = raw_game.split(" ")
        
        rounds = raw_rounds.split("; ")
        for r in rounds:
            cubes = r.split(", ")
            for cube in cubes:
                value, color = cube.split(" ")
                if int(value) > limits[color]:
                    break
            else:
                continue
            break
        else:
            game_sum += int(game_number)

# print(game_sum)
```

```python
# Part One but Better

limits = {
    'red': 12,
    'green': 13,
    'blue': 14,
}

game_sum = 0

with open("2023-12-02_puzzle_input.txt") as f:
    for game in f:
        if game is None:
            continue
        game = game.replace(": ", ", ").replace("; ", ", ").strip()
        cubes = sorted(game.split(", ")[1:], key=lambda x: int(x.split(" ")[0],), reverse=True)
        for cube in cubes:
            value, color = cube.split(" ")
            if int(value) > limits[color]:
                break
            else:
                continue
        else:
            game_sum += int(game.split(", ")[0].split(" ")[1])
            
# print(game_sum)
```

```python
# Part Two
from functools import reduce

sum_min_power = 0

with open("2023-12-02_puzzle_input.txt") as f:
    for game in f:
        if game is None:
            continue
        game = game.replace(": ", ", ").replace("; ", ", ").strip()

        # I think its actually inefficient to sort this
        cubes = sorted(game.split(", ")[1:], key=lambda x: int(x.split(" ")[0],), reverse=True)

        max_values = {
            'red': None,
            'green': None,
            'blue': None,
        }
        
        for cube in cubes:
            value, color = cube.split(" ")
            if max_values[color] == None or int(value) > max_values[color]:
                max_values[color] = int(value)
                if not None in max_values.values():
                    break
    
        sum_min_power += reduce(lambda x, y: x * y, max_values.values())

# print(sum_min_power)
        

```
