```python
# Part 1

import re

digits_re = r"(\d+)"
symbols_re = r"([^\.\d])"

part_number_sum = 0

lines = []
digits = []

def overlap_check(target_idx, start, digit):
    for check in [-1, 0, 1]:
        if target_idx + check < 0:
            continue
        start_buffer = start - 1 if start > 0 else start
        end_buffer = start + len(digit) + 1 if start + len(digit) + 1 < len(lines[target_idx + check]) else start + len(digit)
        if re.search(symbols_re, lines[target_idx + check][start_buffer:end_buffer]):
            return int(digit)
    return 0


with open("2023-12-03_puzzle_input.txt") as f:
    for idx, line in enumerate(f):
        line = line.strip()
        lines.append(line)
        digits.append([(dig.start(), dig.group(0)) for dig in re.finditer(digits_re, line)])

        if idx < 1:
            continue

        target_idx = idx - 1 
        for start, digit in digits[target_idx]:
            part_number_sum += overlap_check(target_idx, start, digit)


# check the last line
target_idx = len(lines) - 1
for start, digit in digits[target_idx]:
    for check in [-1, 0]:
        if re.search(symbols_re, lines[target_idx + check][start - 1:start + len(digit) + 1]):
            part_number_sum += int(digit)
            break

# print(part_number_sum)
```
